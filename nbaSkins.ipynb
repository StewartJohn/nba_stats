```python
import pandas as pd
from nba_api.stats.endpoints import leaguestandings
# leaguestandings
standings = leaguestandings.LeagueStandings(season='2021-22')
standings_df = standings.get_data_frames()[0]
```

```python
#Reset the dataframe index to team name to make it easier to look up wins and losses
standings_dup = standings_df
standings_dup.set_index(['TeamName'], inplace = True)
standings_dup
```

```python
#How to find a teams wins or losses. WINS and LOSSES are both columns that are obscured among the 80 columns in the dataframe
print('Pistons won ', standings_dup.at['Pistons', 'WINS'], ' and lost ', standings_dup.at['Pistons', 'LOSSES'])
```

```python
#Define one of our players scores by adding all of their pickes together
ryan_score = standings_dup.at['Pistons', 'LOSSES'] + standings_dup.at['76ers', 'WINS']
print("Ryan has ", ryan_score, " points.")
```
